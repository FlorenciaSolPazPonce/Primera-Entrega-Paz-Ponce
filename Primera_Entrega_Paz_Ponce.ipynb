# **Procesamiento de datos**

ReemplacÃ© los valores de "Ninguno" en las columnas Dias_Enfermo y Mala_Salud_Mental/Dias por 0 porque para poder generar el modelo de machine learning necesito tener todos los valores en forma numÃ©rica

```python
diab['Dias_enfermo'] = diab['Dias_enfermo'].replace('Ninguno', 0)
diab['Dias_enfermo'] = diab['Dias_enfermo'].astype(int)
```

```python
diab['Mala_Salud_Mental/Dias'] = diab['Mala_Salud_Mental/Dias'].replace('Ninguno', 0)
diab['Mala_Salud_Mental/Dias'] = diab['Mala_Salud_Mental/Dias'].astype(int)
```

Traemos todas las columnas cambiadas a forma numÃ©rica.

```python
pd.get_dummies(diab, drop_first=True)
```

```python
diab_procesado = pd.get_dummies(diab, drop_first=True)
```

```python
diab_procesado
```

```python
columnas_a_eliminar = ["DiabetesColumna", "Diabetes_Si"]
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

# Modelado del Arbol de decisiÃ³n

```python
tree = DecisionTreeClassifier(max_depth=None)
```

DefinÃ­ x e y de nuestro modelo

```python
X = diab_procesado.drop(columnas_a_eliminar, axis=1)
y = diab_procesado[columnas_a_eliminar]
```

```python
X
```

Una vez ya definidas x e y procedemos al anÃ¡lisis del arbol de deicisiÃ³n y posteriormente la realizaciÃ³n del grÃ¡fico.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier(max_depth=5)
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
